```python
#import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import matplotlib.pyplot as plt
import folium
from folium.plugins import FastMarkerCluster, HeatMap
import pandas as pd
```

## Geocoding a single address

```python
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode("India Gate, Delhi, India")
```

```python
location.address
```

```python
lat,long = location.latitude, location.longitude
lat,long
```

```python
#created a map object
map1 = folium.Map(
    location=[lat,long],
    tiles='cartodbpositron',
    zoom_start=20,     #how to zoom in zoom out cirxle
)
folium.CircleMarker(location=[lat,long]).add_to(map1)    #place circle at long,lat
```

```python
map1
```

## Geocoding multiple address

```python
df = pd.read_csv("zomato.csv", encoding='latin-1')
```

```python
df.head()
```

```python
df_s = df.sample(50)
```

```python
df_s
```

```python
#folium Map
map2 = folium.Map(
    location=[28.556828,77.174210],
    tiles='cartodbpositron',
    zoom_start=12,
)

#making circle for every hotel
df_s.apply(lambda row:folium.CircleMarker(location=[row["Latitude"], row["Longitude"]]).add_to(map2), axis=1)
map2
```

```python
df_s = df.sample(500)
```

```python
map2 = folium.Map(
    location=[28.556828,77.174210],
    tiles='cartodbpositron',
    zoom_start=12,
)
df_s.apply(lambda row:folium.CircleMarker(location=[row["Latitude"], row["Longitude"]]).add_to(map2), axis=1)
map2
```

## Plot Clustered points 

```python
folium_map = folium.Map(location=[28.556828,77.174210],
                        zoom_start=12,
                        tiles='cartodbpositron')


FastMarkerCluster(data=list(zip(df_s['Latitude'].values, df_s['Longitude'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)
folium_map
```
