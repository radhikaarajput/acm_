```python
#matplotlib ---> lib for visualization in python
#get necessary libraries
import matplotlib.pyplot as plt

```

```python
# get the data ready

x = [0,1,2,3,4,5]           #y=x*2
y = [0,1,4,9,16,25]

#we have 2 bins, and each bin will conatin 12.5(25/2)
#bin1=[]
#bin2=[]

#x aix-> frequency
#y-> no of buckets

#to make evenly distribute--> 25or max than 25 here

# a simple graph --->telling matplotlib to plot a graph
plt.plot(x, y)




# show the graph
plt.show()
```

```python
# make it a bar plot
plt.bar(x, y)
plt.show()
```

```python
# change the color of bars
plt.bar(x, y, color='red')
plt.show()
```

```python
# make it horizontal bar---> barh
plt.barh(x, y, color='yellow')
plt.show()
```

```python
# make a scatter plot---> just put points--> gives bigger view
#it shows density and overall view of data

plt.scatter(x, y)
plt.show()
```

```python
# make those histograms
#histogram->shows frequency 

plt.hist(x, y)
plt.show()
```

```python
# make a proper histograms 
#bins-->make data evenly distribute

plt.hist(y, bins=7)
plt.show()
```

```python
plt.hist(y, bins=2)
plt.show()
```

```python
plt.hist(y, bins=1)
plt.show()
```

```python
# do some real data [kind of]
#contiuous data

import numpy as np

#linspace---> give 100 numbers b/w 0-10 on equal interval
x = np.linspace(0, 10, 100) 
print(x)
```

```python
plt.plot(np.sin(x))
plt.plot(np.cos(x))    #do plot on top of the previous tab and change the default colour to blue
# as 2 plot called back to back and then show
#blue->cos
#red->sin
plt.show()
```

```python
# do some real data [kind of]

import numpy as np

#continuous data
x = np.linspace(0, 10, 100)

plt.plot(np.tan(x))
plt.plot(np.arctan(x))       
plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.show()
```

```python
# make this look better

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), label='tan')
plt.plot(np.arctan(x), label='arctan')
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.ylim(-2, 2)   #limiting y axis from -2 to 2 so that we have a better look at previous 
plt.show()
```

```python
# give label to all these now use 'label'

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), label='tan')
plt.plot(np.arctan(x), label='arctan')
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.ylim(-2, 2)
plt.legend()
plt.show()
```

```python
# change dash styles
# make this look better

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), '-', label='tan')  #here line of ----
plt.plot(np.arctan(x), '--', label='arctan')   #line-- -- --
plt.plot(np.sin(x), 'x', label='sin')
plt.plot(np.cos(x), '*', label='cos')
plt.ylim(-2, 2)
plt.legend()
plt.show()
```
