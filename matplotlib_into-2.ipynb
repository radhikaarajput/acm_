```python
#seaborn library-- for making data more presentable 

#get necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# get the data ready

x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]


# a simple graph 
plt.plot(x, y)


# show the graph
plt.show()
```

```python
# make it a bar plot
plt.bar(x, y)
plt.show()
```

```python
# change the color of bars
plt.bar(x, y, color='red')
plt.show()
```

```python
# make it horizontal bar
plt.barh(x, y, color='yellow')

#enmueraet get a index and value (y gives value)
for index,value in enumerate(y):
    print(index,value)      
    plt.text(value,index,str(value))  #str(value for putting text
    
    
plt.show()
```

```python
# make vertical bar
plt.bar(x, y, color='blue')

#enmueraet get a index and value (y gives value)
for index,value in enumerate(y):
    print(index,value)      
    plt.text(index,value,str(value))  #str(value for putting text
    
    
plt.show()
```

```python
# make a scatter plot

plt.scatter(x, y)
plt.show()
```

```python
# make those histograms 

plt.hist(x, y)
plt.show()
```

```python
# make a proper histograms 

plt.hist(y, bins=7)
plt.show()
```

```python
# do some real data [kind of]

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.show()
```

```python
# do some real data [kind of]

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x))
plt.plot(np.arctan(x))
plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.show()
```

```python
# make this look better

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), label='tan')
plt.plot(np.arctan(x), label='arctan')
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.ylim(-2, 2)
plt.show()
```

```python
# lets give a legend

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), label='tan')
plt.plot(np.arctan(x), label='arctan')
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.ylim(-2, 2)
plt.legend()
plt.show()
```

```python
# change dash styles
# make this look better

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(np.tan(x), '-', label='tan')
plt.plot(np.arctan(x), '--', label='arctan')
plt.plot(np.sin(x), 'x', label='sin')
plt.plot(np.cos(x), '*', label='cos')
plt.ylim(-2, 2)
plt.legend()
plt.show()
```

```python
# generate some random data
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
plt.show()
```

```python
import pandas as pd

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col])
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
plt.show()
```

```python
sns.kdeplot(data)
plt.show()
```

```python
sns.jointplot('x', 'y', data)
plt.show()
```

```python
sns.jointplot('x', 'y', data, kind='hex')
plt.show()
```

```python
sns.jointplot('x', 'y', data, kind='kde')
plt.show()
```
